```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.special import factorial
```

```python
# Definición Poisson
def poisson(n, lamda):
    return lamda ** n * np.exp(-lamda) / factorial(n)

# Función de probabilidad condicional
def prob(n, lamda):
    if n == 0:
        return 0
    else:
        return poisson(n, lamda) / (1 - poisson(0, lamda))  # probabilidad condicional dada
#valor lamda
lamda = 50/60 #cincuenta particulas por minuto

# Valores de n
n_valores = np.arange(1, 11) #un array de 1 a 10 valores
prob_conditional = [prob(n, lamda) for n in n_valores]

#grafico
plt.figure(figsize=(8, 6))
plt.bar(n_valores, prob_conditional, edgecolor='black')
plt.xlabel('Número de partículas (n)')
plt.ylabel('Probabilidad condicional P(N = n | N ≥ 1)')
plt.title('Distribución de probabilidad condicional de impactos de partículas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
